```python
import scipy.stats as stats
import pandas as pd

contingency = pd.crosstab(df['medical_condition'], df['recommended_diet'])
chi2, p, dof, ex = stats.chi2_contingency(contingency)
print(p)
```

```python
import numpy as np

def cramers_v(confusion_matrix):
    chi2 = stats.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr/min((kcorr-1),(rcorr-1)))

contingency = pd.crosstab(df['diet_preference'], df['recommended_diet'])
contingency1 = pd.crosstab(df['goal'], df['recommended_diet'])
contingency2 = pd.crosstab(df['medical_condition'], df['recommended_diet'])

print(cramers_v(contingency))
print(cramers_v(contingency1))
print(cramers_v(contingency2))
```

```python
#Anova(Numerical vs Categorical)
from scipy.stats import f_oneway
annova_results = {}

for num_col in df[num_cols]:
    groups = [group[[num_col]].values 
                for name, group in df.groupby("recommended_diet")]
    f_stat, p_val = f_oneway(*groups)
    annova_results[num_col] = {"F-stat" : f_stat, "p-value": p_val}

annova_results
```

**It's time to fix the outliers in the numerical columms!**

```python
def fix_outliers(col):
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR

    med = df[col].median()  # FIX: call the method

    df.loc[(df[col] < lower) | (df[col] > upper), col] = med

for col in num_cols:
    fix_outliers(col)
```

```python
import math

cols = df[num_cols].columns
n = len(cols)
rows = math.ceil(n / 3)

plt.figure(figsize=(15, 4 * rows))

for i, col in enumerate(cols, 1):
    plt.subplot(rows, 3, i)
    plt.boxplot(df[col])
    plt.title(col)

plt.tight_layout()
save_fig("Outliers fixed plot")
plt.show()
```
